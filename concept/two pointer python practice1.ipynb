```python
def square_list(nums : list)->list:
    left,right=0,len(nums)-1
    result=[]
    while left<= right:
        if abs(nums[left])>abs(nums[right]):
            result.append(nums[left]**2)
            left+=1
        else:
            result.append(nums[right]**2)
            right-=1
    return result[::-1]
print(square_list([-4,-1,0,3,10]))
print(square_list([-9,-1,0,3,7]))
```

```python
class Solution:
    def threesum(self, num: list[int])-> list[list[int]]:
        num.sort()
        result=[]
        n=len(num)
        target=0
        for i in range(n):
            if num[i]>0:
                break
            if i>0 and num[i]==num[i-1]:
                continue    
            low,high=i+1,n-1
            while low<high:
                sum=num[i]+num[low]+num[high]
                if sum==target:
                    result.append([num[i],num[low],num[high]]) 
                    low,high=low+1,high-1
                    while low<high and num[low]==num[low-1]:
                        low+=1
                    while low<high and num[high]==num[high+1]:
                        high-=1
                elif sum<target:
                    low+=1
                else:
                    high-=1
        return result

s=Solution()
print(s.threesum([-1,0,0,1,1,2,-1,-4]))
print(s.threesum([2,-1,3,5,-3,-2,0,-1,-2,-4,1]))
print(s.threesum([2,1,9,6,7,5,2,7,10]))
print(s.threesum([-1,-1,0,0,1,1,2,-1,-4]))


```

```python
class Solution:
    def ispalindrome(self, s:str)-> bool:
        n=len(s)
        l,r=0,n-1
        while l<r:
            if not s[l].isalnum():
                l+=1
                continue
            if not s[r].isalnum():
                r-=1
                continue
            if s[l].lower()!=s[r].lower():
                return False
            l+=1
            r-=1
        return True
s=Solution()
print(s.ispalindrome("A man, a plan, a canal: Panama"))
print(s.ispalindrome("race a car"))
print(s.ispalindrome(" "))
print(s.ispalindrome("madam"))
```

```python
class solution():
    def threeSumClosest(self, nums: list[int], target: int) -> int:
        nums.sort()
        l=0
        n=len(nums)
        closest=float('inf')
        for i in range(n):
            if i>0 and nums[i]==nums[i-1]:
                continue
            l=i+1
            r=n-1
            while l<r:
                sum=nums[i]+nums[l]+nums[r]
                if abs(target-sum)<abs(target-closest):
                    closest=sum
                if sum==target:
                    return sum
                elif sum<target:
                    l+=1
                else:
                    r-=1
        return closest
s=solution()
print(s.threeSumClosest([-1,2,1,-4],1))
print(s.threeSumClosest([0,0,0,0],1))
print(s.threeSumClosest([0,0,0,2,1],3))
```

```python
class Solution:
    def removeduplicate(self, nums: list[int]) -> list[int]:
        if not nums:
            return nums

        nums.sort()
        i = 0

        for j in range(1, len(nums)):
            if nums[i] != nums[j]:
                i += 1
                nums[i] = nums[j]

        return nums[:i + 1]  # +1 because index starts at 0


s = Solution()
print(s.removeduplicate([3, 2, 2, 8, 3]))
print(s.removeduplicate([2, 1, 2, 2, 3, 0, 4, 2, 7]))
print(s.removeduplicate([2, 3, 5, 4, 7, 7]))
```

```python
class Solution:
    def numSubseq(self, nums: list[int], target: int) -> int:
        # Step 1: Sort the array
        nums.sort()
        
        # Step 2: Initialize two pointers
        left, right = 0, len(nums) - 1
        result = 0
        
        # Step 3: Traverse the array using two pointers
        while left <= right:
            if nums[left] + nums[right] <= target:
                # If nums[left] + nums[right] <= target,
                # the number of valid subsequences is 2^(right - left)
                result += pow(2, right - left)
                left += 1  # Move left pointer to the right
            else:
                right -= 1  # Move right pointer to the left
        
        return result
s=Solution()
print(s.numSubseq([3,5,6,7],9))
print(s.numSubseq([3,3,6,8],10))
print(s.numSubseq([2,3,3,4,6,7],12))
```

```python
class Solution:
    def longsubstrwithoutrep(self, s: str) -> int:
        l = 0  # Left pointer of the sliding window
        longest = 0  # Variable to store the length of the longest substring
        seen = set()  # Set to track characters in the current window

        for r in range(len(s)):  # Right pointer of the sliding window
            while s[r] in seen:  # If character at 'r' is already in the window
                seen.remove(s[l])  # Remove character at 'l' from the window
                l += 1  # Shrink the window from the left
            seen.add(s[r])  # Add character at 'r' to the window
            longest = max(longest, r - l + 1)  # Update the longest length
        return longest
s = Solution()
print(s.longsubstrwithoutrep("abcabcbb"))  # Output: 3
print(s.longsubstrwithoutrep("bbbbb"))     # Output: 1
print(s.longsubstrwithoutrep("pwwkew"))    # Output: 3
print(s.longsubstrwithoutrep("dvdf"))      # Output: 3
```

```python
class Solution:
    def maxaversubarray(self, nums: list[int], k: int) -> float:
        window_sum = 0
        n=len(nums)

        for i in range(k):
            window_sum+=nums[i]
        max_avg=window_sum/k
        for i in range(k,n):
            window_sum-=nums[i-k]
            window_sum+=nums[i]
            max_avg=max(max_avg,window_sum/k)
        return max_avg
s=Solution()            
print(s.maxaversubarray([1,12,-5,-6,50,3],4))
print(s.maxaversubarray([5],1))
```

```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        counts = {}           # Tracks frequency of each character in window
        max_count = 0         # Highest frequency of any single character in window
        left = 0
        longest = 0

        for right, char in enumerate(s):
            counts[char] = counts.get(char, 0) + 1
            max_count = max(max_count, counts[char])

            # Shrink window if too many replacements needed
            if (right - left + 1) - max_count > k:
                counts[s[left]] -= 1
                left += 1

            longest = max(longest, right - left + 1)

        return longest
s=Solution()
print(s.characterReplacement("ABAB",2))
print(s.characterReplacement("AABCDBBA",1))
print(s.characterReplacement("DCDEERT",1))
```

```python
#binary search
class Solution:
    def search(self, nums: list[int], target: int) -> int:
        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = (left + right)//2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        return -1
s=Solution()
print(s.search([-1,0,3,5,9,12],9))
print(s.search([-1,0,3,5,9,12],2))
print(s.search([5],5))
```

```python
#mergesort
class Solution:
    def merge(self, left: list[int], right: list[int]) -> list[int]:
        result = []
        i = j = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1
        result.extend(left[i:])
        result.extend(right[j:])
        return result

    def mergeSort(self, nums: list[int]) -> list[int]:
        if len(nums) <= 1:
            return nums
        mid = len(nums) // 2
        left = self.mergeSort(nums[:mid])
        right = self.mergeSort(nums[mid:])
        return self.merge(left, right)

s = Solution()
print(s.mergeSort([5,2,3,1]))
print(s.mergeSort([5,1,1,2,0,0]))
```

```python
# Iterative merge sort (non-recursive)
class Solution:
    def mergeSort(self, nums: list[int]) -> list[int]:
        subarray_size = 1
        n = len(nums)
        result = nums[:]
        while subarray_size < n:
            for i in range(0, n, 2 * subarray_size):
                left = result[i:i+subarray_size]
                right = result[i+subarray_size:i+2*subarray_size]
                merged = []
                l = r = 0
                while l < len(left) and r < len(right):
                    if left[l] < right[r]:
                        merged.append(left[l])
                        l += 1
                    else:
                        merged.append(right[r])
                        r += 1
                merged.extend(left[l:])
                merged.extend(right[r:])
                result[i:i+2*subarray_size] = merged
            subarray_size *= 2
        return result

s = Solution()
print(s.mergeSort([5,2,3,1]))
print(s.mergeSort([5,1,1,2,0,0]))
```

```python
class Solution:
    def maxProduct(self, nums: list[int]) -> int:
        max_prod = min_prod = result = nums[0]
        for num in nums[1:]:
            candidates = (num, max_prod * num, min_prod * num)
            max_prod = max(candidates)
            min_prod = min(candidates)
            result = max(result, max_prod)
        return result

s = Solution()
print(s.maxProduct([2,3,-2,4]))
print(s.maxProduct([-2,0,-1]))
```

```python
class Queue:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        raise IndexError("dequeue from an empty queue")

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def peek(self):
        if not self.is_empty():
            return self.queue[0]
        raise IndexError("peek from an empty queue")

q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.queue)
print(q.dequeue())
print(q.dequeue())
print(q.queue)
print(q.peek())
print(q.queue)
print(q.size())
print(q.is_empty())
```

```python
class CircularQueue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.queue = [None] * capacity
        self.front = -1
        self.rear = -1
        self.size = 0

    def is_empty(self):
        return self.size == 0

    def is_full(self):
        return self.size == self.capacity

    def enqueue(self, item):
        if self.is_full():
            print("Queue is full. Cannot enqueue.")
            return

        if self.is_empty():
            self.front = 0
            self.rear = 0
        else:
            self.rear = (self.rear + 1) % self.capacity
        
        self.queue[self.rear] = item
        self.size += 1
        print(f"Enqueued: {item}")

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty. Cannot dequeue.")
            return None

        dequeued_item = self.queue[self.front]
        self.queue[self.front] = None  # Optional: clear the dequeued spot
        
        if self.front == self.rear:  # Last element dequeued
            self.front = -1
            self.rear = -1
        else:
            self.front = (self.front + 1) % self.capacity
        
        self.size -= 1
        print(f"Dequeued: {dequeued_item}")
        return dequeued_item

    def peek(self):
        if self.is_empty():
            print("Queue is empty.")
            return None
        return self.queue[self.front]

    def display(self):
        if self.is_empty():
            print("Queue is empty.")
            return

        print("Circular Queue elements:")
        if self.rear >= self.front:
            for i in range(self.front, self.rear + 1):
                print(self.queue[i], end=" ")
        else:
            for i in range(self.front, self.capacity):
                print(self.queue[i], end=" ")
            for i in range(0, self.rear + 1):
                print(self.queue[i], end=" ")
        print()
s=CircularQueue(5)
s.enqueue(2)
print(s.queue)
```
